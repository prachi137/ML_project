#importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#create the data frame

```python
dataframe = pd.read_csv("/content/drive/MyDrive/Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

#convert the datatype of column rate

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

#type of resturant


```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('type of resturant')
```

#conclusion - majority of the resturant falls in dinning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker='o')
plt.xlabel('type of resturant',c='red',size=20)
plt.ylabel('votes',c='red',size=20)
```

#coclusion- dinning resturants has received maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title('ratings distribution')
plt.show()
```

#conclusion- the majority resturant received rating from 3.5 to 4

```python
couple_data=dataframe['approx_cost(for two people)']
sns.distplot(x=couple_data)
```

#which mode receives maximum rating

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate',data = dataframe)
```

#conclusion- offline order received lower rating in comparison to online

#coclusion-dining resturant primarily accept offline order whereas cafes received online order. this suggest that client prefers order in person in restaurants,but pefer online ordering at cafe

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True,cmap='YlGnBu',fmt='d')
plt.title("heatmap")
plt.xlabel("online_order")
plt.ylabel(("Listed In(Type)"))
plt.show()
```
