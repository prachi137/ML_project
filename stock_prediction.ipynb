```python
import pandas as pd
df.reset_index(drop=True, inplace=True)
lag_feature = ['Open','High','Low','Close']
window1 = 3
window2 = 7
df_rolled_3d = df[lag_feature].rolling(window=window1, min_periods=0)
df_mean_3d = df_rolled_3d.mean().shift(1).reset_index().astype(np.float32)

df_rolled_7d = df[lag_feature].rolling(window=window2, min_periods=0)
df_mean_7d = df_rolled_7d.mean().shift(1).reset_index().astype(np.float32)

df_std_7d = df_rolled_7d.std().shift(1).reset_index().astype(np.float32)
for feature in lag_feature:
  df[f"{feature}_mean_lag{window1}"] = df_mean_3d[feature].astype('float32')
  df[f"{feature}_std_lag{window2}"] = df_std_7d[feature].astype('float32')
```

```python
import pandas as pd
import numpy as np


df['Date'] = pd.to_datetime(df['Date'])


numeric_columns = df.select_dtypes(include=np.number).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

**sacalling data and transforming it for training**

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

y_traiyn = df_train['Close']
scaler = MinMaxScaler(feature_range=(0, 1))
y_train1 = scaler.fit_transform(np.array(df_train['Close']).reshape(-1, 1))
y_trainn = df_train['Close'].to_numpy().reshape(-1, 1)
```

```python
y_validn = df_train['Close']
scaler = MinMaxScaler(feature_range=(0, 1))
y_valid1 = scaler.fit_transform(np.array(df_train['Close']).reshape(-1, 1))
y_validn = df_train['Close'].to_numpy().reshape(-1, 1)
```

```python
x_train = df_train.drop(columns=['Close'])

x_train1 = x_train.drop('Date', axis=1)

x_train1 = (x_train1 - x_train1.min()) / (x_train1.max() - x_train1.min())


x_train2 = pd.concat([x_train1, x_train.Date], axis=1)
```

```python
x_valid = df_train.drop(columns=['Close'])

x_valid1 = x_train.drop('Date', axis=1)

x_valid1 = (x_train1 - x_train1.min()) / (x_train1.max() - x_train1.min())


x_valid2 = pd.concat([x_train1, x_train.Date], axis=1)
```

```python
x_train3 = x_train2.drop(columns=['Date']).to_numpy()
x_valid3 = x_valid2.drop(columns=['Date']).to_numpy()
```

```python
y_train1 = y_train1.reshape(-1,1)
y_valid1 = y_valid1.reshape(-1,1)
```

```python
!pip install xgboost
import xgboost as xgb
import sklearn
from xgboost import XGBRegressor
from xgboost import plot_importance
```

**model training**

```python
xgb = XGBRegressor(n_estimators=1000, learning_rate=0.01)
xgb

xgb.fit(x_train3, y_train1, eval_set=[(x_train3,y_train1),(x_valid3, y_valid1)],early_stopping_rounds=100, verbose=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
feature_importance = plot_importance(xgb, height=0.9)
feature_importance
```

```python
predicted_results_v = xgb.predict(x_valid3)
predicted_results_t = xgb.predict(x_train3)
predicted_results_v = predicted_results_v.reshape(-1,1)
predicted_results_t = predicted_results_t.reshape(-1,1)
predicted_results_v = scaler.inverse_transform(predicted_results_v)
predicted_results_t = scaler.inverse_transform(predicted_results_t)
```

```python
plt.figure(figsize=(20,15))
plt.plot(x_train2.index,list(y_trainn),label="actual Price train",color='blue')
plt.plot(x_train2.index,list(predicted_results_t),label="pridicted Price train",color='red')
plt.plot(x_train2.index,list(y_validn),label="actual Price valid",color='maroon')
plt.plot(x_train2.index,list(predicted_results_v),label="predicted  Price valid",color='green')

plt.xlabel('Date')
plt.ylabel('Nifty so closing Price')
plt.legend()
plt.show()


```
